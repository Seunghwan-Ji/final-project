```python
# 재료 번역 딕셔너리 파일 불러오기

import json

# JSON 파일에서 딕셔너리 읽어오기
with open('mapping_dict.json', 'r', encoding='utf-8') as file:
    mapping_dict = json.load(file)

mapping_dict
```

```python
# 인식된 재료 예시
detected_li = ['감자', '계란', '당근']

# 재료 열에서 각 값을 람다함수 적용, 값은 x에 해당됨, any(): 위에서 인식된 재료 리스트중 하나라도 x에 포함되어있으면 true 반환
filtered_df = df[df['CKG_MTRL_CN'].apply(lambda x: any(item in str(x) for item in detected_li))]

# 추출된 내용 확인
filtered_df.head()
```

```python
# 재료 리스트
detected_li = ['감자', '계란', '당근']

# 모든 재료를 포함하는 행 필터링, all(): 위에서 인식된 재료 리스트 전부가 x에 포함되어있으면 true 반환
filtered_df = df[df['CKG_MTRL_CN'].apply(lambda x: all(item in str(x) for item in detected_li))]

# 결과 확인
filtered_df.head()
```
