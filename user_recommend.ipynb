```python
import re

# 재료이름만 남기는 함수(예외 존재)
def remain_food_name(food):
    # food가 str 타입이 아니면 빈문자열 반환
    if not isinstance(food, str):
        return ""
    
    # 정규식 패턴
    pattern = r'\[\w+\]'

    # 정규식으로 '[한글]' 형식의 부분을 찾아 분리
    parts = re.split(pattern, food)
    
    # [한글] 형식 제외한 부분 저장
    parts = [part.strip() for part in parts]
    
    # '| ' 부분을 제거해서 재료들을 각각 result에 저장
    result = []
    for i in parts:
        result.extend(i.split('| '))
        
    return ', '.join(result[1:])
```

```python
df['재료리스트2'] = df['재료리스트'].apply(remain_food_name)
df['재료리스트2']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# CountVectorizer : 자연어처리에서 텍스트 데이터를 숫자형 데이터로 변환하는데 사용
# input: 입력 데이터의 형식을 지정합니다. (기본값: 'content')
# encoding: 입력 데이터의 인코딩을 지정합니다. (기본값: 'utf-8')
# decode_error: 디코딩 에러 처리 방법을 설정합니다. (기본값: 'strict')
# stop_words: 불용어(stop words)를 제거할지 여부를 설정합니다. (예: 'english'는 영어 불용어를 제거함)
# ngram_range: n-그램의 범위를 지정합니다. (예: (1, 2)는 단어와 두 단어 조합을 모두 포함함)
# max_features: 가장 빈도가 높은 상위 n개의 단어만을 사용하도록 설정합니다.
# tokenizer: 단어를 토큰화하는 함수를 지정합니다. (기본값: None)
# preprocessor: 전처리 함수를 지정합니다. (기본값: None)
# analyzer: 분석 방법을 설정합니다. ('word'는 단어 단위, 'char'는 문자 단위)
# max_df: 특정 단어가 전체 문서에서 나타나는 비율의 최대값을 설정합니다. (예: 0.5는 50% 이상 나타나는 단어는 제외)
# min_df: 특정 단어가 전체 문서에서 나타나는 최소 빈도수를 설정합니다. (예: 2는 최소 2개 문서에 나타나는 단어만 사용)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 텍스트 벡터화
text_vectorizer = CountVectorizer()

# 범주형 인코딩
categorical_encoder = OneHotEncoder(handle_unknown='ignore') # 존재하지 않는 새로운 범주가 데이터에 나타나면 새로운 범주는 무시

# ColumnTransformer를 사용하여 텍스트와 범주형 변수 처리
preprocessor = ColumnTransformer(
    transformers=[
        ('text', text_vectorizer, '재료리스트2'),
        ('cat', categorical_encoder, ['요리명', '요리타입', '음식분위기', '재료타입', '음식타입', '몇인분', '요리난이도', '요리시간'])
    ]
)

# 데이터 전처리 및 벡터화
X = preprocessor.fit_transform(recommend_df)

# 사용자 선호 레시피 (예: 0번 행을 시킨 경우)
user_recipe_index = 0 # 어묵김말이
user_recipe_vec = X[user_recipe_index]

# 유사도 계산
cosine_similarities = cosine_similarity(user_recipe_vec.reshape(1, -1), X).flatten()

# 유사도에 따라 레시피 추천 (자기 자신 제외)
similar_recipe_indices = cosine_similarities.argsort()[::-1][1:6]  # 유사도가 높은 상위 5개 레시피 추천

recommended_recipes = recommend_df.iloc[similar_recipe_indices]

recommended_recipes
```
